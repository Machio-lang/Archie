```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path= r"C:\Users\k00009513\Downloads\KRA_Class_Dataset.xlsx"
data=pd.read_excel(path)
data.head(5)
```

```python
plt.bar(data['sector'], data['tax_paid'])
plt.xlabel("sector")
plt.ylabel("tax_paid")
plt.title("Vertical Bar Chart")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
data.groupby('sector')['tax_paid'].sum().sort_values().plot(kind='bar', color='skyblue')
plt.title('Total Tax Paid per Sector')
plt.xlabel('Sector')
plt.ylabel('Total Tax Paid')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(6, 6))
data['compliance_status'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['lightcoral','grey', 'lightblue'])
plt.title('Proportion of Taxpayer Types')
plt.ylabel('')  # Remove y-axis label
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(data['tax_paid'], bins=20, edgecolor='black', color='blue')
plt.title('Distribution of Tax Paid')
plt.xlabel('Tax Paid')
plt.ylabel('Frequency')
plt.show()
```
